```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import july

plt.style.use('seaborn')
plt.rcParams['figure.facecolor'] = 'none'
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
Columns = ['DayofMonth','DayOfWeek','FlightDate','Marketing_Airline_Network','Operating_Airline ','Origin',
          'OriginCityName','OriginStateName','Dest','DestCityName','DestStateName','CRSDepTime','DepTime','DepDelay','DepDelayMinutes','TaxiOut','TaxiIn',
          'CRSArrTime','ArrTime','ArrDelay','ArrDelayMinutes','Cancelled','Diverted','CRSElapsedTime',
          'ActualElapsedTime','AirTime','Distance','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']
df = pd.read_csv('data.csv',usecols=Columns)
df.head()
```

```python
df.info(verbose=True)
```

## Columns Preprocessing

```python
# data preprocessing
numerical_data =  [var for var in df.columns if df[var].dtype!='O'] 
categorial_data = [var for var in df.columns if df[var].dtype =='O']

df[categorial_data] = df[categorial_data].astype('category')
```

```python
df.info()
```

```python
# fixing columns name 
df['Operating_Airline'] = df['Operating_Airline ']
```

```python
airline_mapping = {
    'WN': 'Southwest Airlines',
    'DL': 'Delta Air Lines',
    'AA': 'American Airlines',
    'UA': 'United Airlines',
    'OO': 'SkyWest Airlines',
    'YX': 'Republic Airways',
    'B6': 'JetBlue Airways',
    'NK': 'Spirit Airlines',
    'AS': 'Alaska Airlines',
    'MQ': 'Envoy Air',
    'OH': 'PSA Airlines',
    '9E': 'Endeavor Air',
    'F9': 'Frontier Airlines',
    'G4': 'Allegiant Air',
    'PT': 'Piedmont Airlines',
    'YV': 'Mesa Airlines',
    'HA': 'Hawaiian Airlines',
    'C5': 'CommutAir',
    'QX': 'Horizon Air',
    'ZW': 'Air Wisconsin',
    'G7': 'GoJet Airlines'
}

df['Operating_Airline'] = df['Operating_Airline'].map(airline_mapping)

df['Operating_Airline'].value_counts()
```

```python
for i in df.columns:
    temp = df[i].isna().value_counts()
    true_val = temp[temp.index == True]
    if len(true_val) > 0:
        print(i," - Number of missing data :",true_val[0])
```

```python
df.loc[df['ArrTime'].isna()] #ArrTime missing when the flight are cancelled
df['Cancelled'].value_counts()
```

```python
df['DepDelayMinutes']
```

```python
Condition = [df['DepDelayMinutes']==0,
            (df['DepDelayMinutes']>0) & (df['DepDelayMinutes']<=15),
            (df['DepDelayMinutes']>15) & (df['DepDelayMinutes']<=45),
            df['DepDelayMinutes']>45,
            df['Cancelled']==1]

Values = ['On-Time','Short Delay','Medium Delay','Long Delay','Cancelled']

df['FlightStatus'] = np.select(Condition,Values)
```

## Analysing the data

```python
data = df['FlightStatus'].value_counts(normalize=True)

labels = data.keys()
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
fig = plt.figure()

plt.pie(df['FlightStatus'].value_counts(normalize=True), labels=labels, autopct='%1.1f%%',colors=colors, startangle=90,
        wedgeprops={'linewidth': 1, 'edgecolor': 'white'}, textprops={'fontsize': 10},
        shadow=True)
plt.axis('equal')

plt.title('Flight Status',fontsize=15, pad=30)
plt.show()
#save the figure
fig.savefig('Flight Status', transparent=True)
```

```python

# Create the stacked bar plot
ax = df['Operating_Airline'].value_counts(ascending=True).plot(
    kind='barh',
    stacked=True,
    title='Number of flight by Operating Airline',
    mark_right=True,
    figsize=(14,8)  # Adjust the figure size as per your preference
)
# Add labels to the bars
for patch in ax.patches:
    width = patch.get_width()
    ax.annotate(f'{int(width)}', (width, patch.get_y() + patch.get_height() / 2),
                xytext=(20, 0), textcoords='offset points', ha='center', va='center')




# Adjust the layout
plt.tight_layout()

plt.savefig('Number of flight by Operating Airline', transparent=True)

# Display the plot
plt.show()
```

```python
grouped = df.groupby(['Operating_Airline','FlightStatus']).size()
```

```python
normalized_table = grouped.unstack().apply(lambda x: x / x.sum() *100, axis=1)
normalized_table.style.background_gradient(cmap='YlOrRd')
```

```python
columns_order = ['On-Time','Short Delay','Medium Delay','Long Delay','Cancelled']
normalized_table = normalized_table.reindex(columns=columns_order)
sorted_table = normalized_table.sort_values(by='On-Time', ascending=True)


# Create the stacked bar plot
ax = sorted_table.plot(
    kind='barh',
    stacked=True,
    title='Flight Status by Operating Airline',
    mark_right=True,
    figsize=(10, 6)  # Adjust the figure size as per your preference
)


# Move the legend outside the plot
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))


# Adjust the layout
plt.tight_layout()

plt.savefig('Flight Status by Operating Airline', transparent=True)

# Display the plot
plt.show()

```

## Let Analyse the reason of delay

We focus on the flights that have medium or long delay

```python
df['FlightStatus'].value_counts()
```

```python
df['Operating_Airline'].value_counts()
```

```python
df.loc[(df['FlightStatus'] == 'Medium Delay') | (df['FlightStatus'] == 'Long Delay')]
```

```python
#Airline that have the largest number of delay flights
Delay_flight = df.loc[(df['FlightStatus'] == 'Medium Delay') | (df['FlightStatus'] == 'Long Delay')]
Delay_flight['Operating_Airline'].value_counts()
```

```python
Delay_flight['FlightDate'].unique()
```

```python
july.month_plot(Delay_flight['FlightDate'].unique(), Delay_flight['FlightDate'].value_counts(), 
                month=4, date_label=True,title='Delay flight by days',cmap='YlOrRd')
```

```python
delay_columns = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
Delay_flight[delay_columns].sum()
```

```python
for j,i in Delay_flight[delay_columns].sum().sort_values(ascending=False).items():
    days, remaining_minutes = divmod(i, 1440)
    hours, minutes = divmod(remaining_minutes, 60)
    formatted_time = f"{int(days)} days, {int(hours):02d}:{int(minutes):02d}"
    print(j," : ",formatted_time)
```

```python
CountFlights = Delay_flight['Operating_Airline'].value_counts()
DelaySum = Delay_flight.groupby('Operating_Airline')['CarrierDelay','LateAircraftDelay'].sum()
DelaySum['TotalDelay'] = DelaySum['CarrierDelay']+ DelaySum['LateAircraftDelay']
DelaySum = DelaySum.reset_index()
```

```python
DelaySum['DelayAverage'] = DelaySum['TotalDelay'] / CountFlights[DelaySum['Operating_Airline']].values
```

```python
DelaySum[['Operating_Airline','DelayAverage']].sort_values('DelayAverage',ascending=False).style.background_gradient(cmap='YlOrRd')
```

```python
WeatherDelay = Delay_flight.groupby('Origin')['WeatherDelay'].mean()
WeatherDelay
```

```python
from geopy.geocoders import Nominatim

# Create a geocoder object
geolocator = Nominatim(user_agent="airport_geocoder")
    
# Create an empty dictionary to store the airport coordinates
airport_coordinates = {}

# Iterate over each airport abbreviation
for abbreviation in WeatherDelay.keys():
    location = geolocator.geocode(abbreviation)
    
    if location is not None:
        # Extract the latitude and longitude coordinates
        latitude = location.latitude
        longitude = location.longitude
        
        # Add the coordinates to the dictionary
        airport_coordinates[abbreviation] = (latitude, longitude)
```

```python
import folium

# Create a map centered at the approximate center of the USA
map_origin = folium.Map(location=[39.8283, -98.5795], zoom_start=4)
TopWeatherDelay = WeatherDelay.sort_values(ascending=False).nlargest(10)
# Iterate over each origin abbreviation and delay time
for origin_abbreviation, delay_time in TopWeatherDelay.items():
    
    # Get the latitude and longitude coordinates for the origin abbreviation
    origin_coordinates = airport_coordinates.get(origin_abbreviation)
    
    if origin_coordinates:
        hours, minutes = divmod(delay_time, 60)
        # Add a marker for each origin with a popup showing the delay time
        folium.Marker(location=origin_coordinates, icon=folium.Icon(color='red', icon='info-sign'),
            popup=f"{origin_abbreviation} ,Delay Time: {int(hours):02d}:{int(minutes):02d}").add_to(map_origin)

# Display the map
map_origin
```
